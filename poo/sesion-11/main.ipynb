```python
#Importar pandas como pd
#leer el csv que se encuentra en https://raw.githubusercontent.com/irojasgo/oop_python/main/netflix_titles.csv
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/irojasgo/oop_python/main/netflix_titles.csv")
```

```python
#Dataset de peliculas emitidas en estados unidos

df[df["country"] == "United States"]
```

```python
#Cantidad de shows por rating

df["rating"].value_counts()
```

```python
#Cantidad de shows por pais

df["country"].value_counts()
```

```python
#Cuantas peliculas salieron en Mexico en el año 2021

df[(df["country"] == "Mexico") & (df["type"] == "Movie") & (df["release_year"] == 2021)]
```

```python
#Cual es el rating mas comun de estas peliculas

df[(df["country"] == "Mexico") & (df["type"] == "Movie") & (df["release_year"] == 2021)]["rating"].value_counts()
```

```python
#EL mes donde mas peliculas salieron en estados unidos
df["date_added"] = pd.to_datetime(df["date_added"])

df["date_added"].dt.month.value_counts()
```

```python
#Quien es la persona en Mexico que mas Shows de standup tiene en Netflix? (usar la columna cst para responder esto, en caso de haber un empate, indicar a todos aquellas personas que tienen mas shows de standup en Mexico)
df[(df["country"] == "Mexico") & (df["listed_in"] == "Stand-Up Comedy")]["cast"].value_counts()
```

```python
#Ordenar los shows de standup en Mexico por duracion, indicar los 3 shows que duran mas tiempo
df[(df["country"] == "Mexico") & (df["listed_in"] == "Stand-Up Comedy")].sort_values(by="duration", ascending=False).head(3)
```

```python
#Extra: Indicar el dia de la semana en que fueron estrenados los 3 shows de standup con mas duracion en Mexico

df[(df["country"] == "Mexico") & (df["listed_in"] == "Stand-Up Comedy")].sort_values(by="duration", ascending=False).head(3)["date_added"].dt.day_name()
```
