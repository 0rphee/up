**5. Abre una imagen a color (de tu elección), imprime sus dimensiones, el número de píxeles y despliega la imagen con `matplotlib`**

```python
img = cv.imread("../OpenCV-images/HappyFish.jpg")[:,:,::-1]

print("(alto, largo, número de canales de color) = ",img.shape)
print("Número de pixeles = ",img.size)

plt.imshow(img)
```

**6. Divide en variables distintas los canales de color de la imagen anterior**

```python
canal_r = img[:,:,0]
canal_g = img[:,:,1]
canal_b = img[:,:,2]

plt.subplot(131)
plt.imshow(canal_r, cmap='gray')
plt.title('Canal R')

plt.subplot(132)
plt.imshow(canal_g, cmap='gray')
plt.title('Canal G')

plt.subplot(133)
plt.imshow(canal_b, cmap='gray')
plt.title('Canal B') 
```

**7. Convierte la imagen de color a escala de grises**

```python
img_gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
print(img_gray.shape)
plt.imshow(img_gray, cmap="gray")
```

**8. Visita cada una de las posiciones de la imagen en escala de grises. Imprime las coordenadas de los píxeles que tengan un valor entre 200 y 255**

```python
# visito pixel a pixel la imagen y muestro su color
filas, columnas = img_gray.shape

val = 0
for i in range(filas):
    for j in range(columnas):
        val = img_gray[i,j]
        if (200 <= val ) & (val <= 255 ):
            print(f"{i},{j} = {img_gray[i,j]}")
```

**9. Crea una imagen morada (R+B) de 300 x 300 y muéstrala**

```python

canal_r  = np.ones((300,300), dtype=np.uint8)*100
canal_g  = np.zeros((300,300), dtype=np.uint8)
canal_b  = np.ones((300,300), dtype=np.uint8)*100

bandera = cv.merge((canal_r,canal_g,canal_b))

plt.imshow(bandera)
```
