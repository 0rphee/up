```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('images/lena.jpg', 0)
```

```python
# Transformaciones morfolgicas
# en imagenes binarizadas
# ej. elemento estructurante, central y 4vecindada
# erosion y dilatacion son las operasiones primitivaso

UMBRAL = 128
img_thresh = cv.threshold(img, UMBRAL, 255, cv.THRESH_BINARY)[1]
plt.imshow(img_thresh, cmap='gray')
# dilacion =
```

```python
s_e = np.ones((3, 3), dtype=np.uint8)  # elemento estructurante
img_erode = cv.erode(img_thresh, s_e, iterations=3)
plt.imshow(np.hstack((img_thresh, img_erode)), cmap='gray')   
plt.show()
```

```python
s_e2 = np.ones((5, 5), dtype=np.uint8)  # elemento estructurante
img_erode = cv.erode(img_thresh, s_e2, iterations=3)
plt.imshow(np.hstack((img_thresh, img_erode)), cmap='gray')   
plt.show()
```

```python
# dilatacion
img_dilate = cv.dilate(img_thresh, s_e, iterations=3)
plt.imshow(np.hstack((img_thresh, img_dilate)), cmap='gray')
plt.show()
```

```python
# dilatacion
img_dilate = cv.dilate(img_thresh, s_e2, iterations=3)
plt.imshow(np.hstack((img_thresh, img_dilate)), cmap='gray')
plt.show()
```
