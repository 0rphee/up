```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('images/lena.jpg', 0)
```

```python
# Transformaciones morfolgicas
# en imagenes binarizadas
# ej. elemento estructurante, central y 4vecindada
# erosion y dilatacion son las operasiones primitivaso

UMBRAL = 128
img_thresh = cv.threshold(img, UMBRAL, 255, cv.THRESH_BINARY)[1]
plt.imshow(img_thresh, cmap='gray')
# dilacion =
```

```python
kernel = np.ones((3, 3), dtype=np.uint8)  # elemento estructurante
img_erode = cv.erode(img_thresh, kernel, iterations=3)
plt.imshow(np.hstack((img_thresh, img_erode)), cmap='gray')   
plt.show()
```

```python
kernel2 = np.ones((5, 5), dtype=np.uint8)  # elemento estructurante
img_erode = cv.erode(img_thresh, kernel2, iterations=3)
plt.imshow(np.hstack((img_thresh, img_erode)), cmap='gray')   
plt.show()
```

```python
# dilatacion
img_dilate = cv.dilate(img_thresh, kernel, iterations=3)
plt.imshow(np.hstack((img_thresh, img_dilate)), cmap='gray')
plt.show()
```

```python
# dilatacion
img_dilate = cv.dilate(img_thresh, kernel2, iterations=3)
plt.imshow(np.hstack((img_thresh, img_dilate)), cmap='gray')
plt.show()
```

```python
# apertura
# primero erosiona y luego dilata
# sirve para remover ruido

img_open = cv.morphologyEx(img_thresh, cv.MORPH_OPEN, kernel)
plt.imshow(np.hstack((img_thresh, img_open)), cmap='gray')
```

```python
# obtener elemento estructurante
kernel3 =  cv.getStructuringElement(cv.MORPH_RECT, (5, 5))
(kernel3)
kernel4 =  cv.getStructuringElement(cv.MORPH_ELLIPSE, (17,17))
(kernel4)
```

```python
img2 = cv.imread('images/blox.jpg', 0)
plt.imshow(img2, cmap='gray')
```

```python
img2_bin = cv.threshold(img2, 127, 255, cv.THRESH_BINARY)[1]
plt.imshow(img2_bin, cmap='gray')
```

```python
plt.imshow(cv.morphologyEx(img2_bin, cv.MORPH_GRADIENT, kernel2), cmap='gray')
```

```python
plt.imshow(cv.morphologyEx(img2_bin, cv.MORPH_OPEN, kernel4), cmap='gray')
```

```python
# elemento estructurante en forma de diagonal
kernel5 = np.array([[0, 0, 1],
                    [0, 1, 0],
                    [1, 0, 0]], dtype=np.uint8)
plt.imshow(cv.morphologyEx(img2_bin, cv.MORPH_OPEN, kernel5), cmap='gray')
```
