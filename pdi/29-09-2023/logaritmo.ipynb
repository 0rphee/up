```python
# transformaciones logartmicas
# usadas para imgenes con bajo contraste y muy oscuras
# s = pixel resul   
# r = pixel original
# c = constante calculada por 255/log(1+m) donde m es el valor mximo de la imagen
img = cv.imread("../images/aloeL.jpg", 0)
# c = 255 / log(1 + np.max(img))
# s = c * log(1 + ) # 1 para evitar log(0)
plt.imshow(img, cmap="gray")
```

```python
s = np.log(1 + img)

# los valores se vuelven an ms claros

plt.imshow(s, cmap="gray")
```

```python
# para que suavizar la funcin, se modifica la constante c

print(s.dtype)
C = 255 / np.log(1 + np.max(img))
print(C)

s = C * np.log(1 + img.astype(np.float32))
print(img.dtype )

plt.imshow(s, cmap="gray")
```
