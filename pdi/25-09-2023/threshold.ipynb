```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# transformaciones de intensidad y filtrado espacial
# funcin threshold
# a partir de un nivel de brillosidad k, a los menores se les asigna 0 y a los mayores 255
# ej. si k=127, los menores a 127 se les asigna 0 y a los mayores 255

img = cv.imread('../images/aloeL.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
# threshold manual
# umbralizacin global
k = 155

filas, columnas = img.shape

img_nueva = np.zeros((filas, columnas), dtype=np.uint8)

for i in range(filas):
    for j in range(columnas):
        val = img[i,j]
        if val < k:
            img_nueva[i,j] = 0
        else:
            img_nueva[i,j] = 255
plt.imshow(img_nueva, cmap='gray')
```

```python
# threshold de opencv
_, nueva_cv = cv.threshold(img, 128, 1, cv.THRESH_BINARY)
plt.imshow(nueva_cv, cmap='gray')
```

```python
# threshold inverso
_, nueva_cv = cv.threshold(img, 128, 1, cv.THRESH_BINARY_INV)
plt.imshow(nueva_cv, cmap='gray')
```

```python
np.unique(nueva_cv)

cv.imwrite('aloeL_threshold.jpg', nueva_cv)
```

```python
# negativos de imgnes
# se obtienee utilizando la funcin de trasnformacin delegativa
# con niveles de intensidad en el rango [0, L - 1]
# r = valor original del pixel
# s = L - 1 - r
# los colores se reemplazan por su color compelementario
# rojo -> cyan
# verde -> magenta
# azul -> amarillo

img = cv.imread('../images/aloeL.jpg', cv.IMREAD_GRAYSCALE)
img_nueva = img.copy()

filas, columnas = img.shape
for i in range(filas):
    for j in range(columnas):
        img_nueva[i,j] = 255 - img[i,j]
plt.imshow(img_nueva, cmap='gray')
```

```python
img_neg = img.copy()
img_neg = 255 - img_neg
plt.imshow(img_neg, cmap='gray')
```

```python
img_neg_np  =np.bitwise_not(img.copy())
plt.imshow(img_neg_np, cmap='gray')
```
