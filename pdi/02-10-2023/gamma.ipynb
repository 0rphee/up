```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('../images/squirrel_cls.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
# transformaciones gamma
# transformaciones de ley de potencia (gamma)

# la formula general es 
# s = c * r^gamma
# elevar el pixel a una potencia gamma por una constante C que suaviza el resultado
# gamma = 1 nos da la identidad

# la corrección gamma es importante para mostrar correctamente las imágnes en uan pantalla, 
# para evitar el blanquo u oscurecimiento de las imágenes cuando se ven desde diferentes tipos de monitores
# estot es pues nosotros vemos en una escala logaritmica y las pantallas son lineales

img_gamma = img.copy()

gamma = 0.05

filas, columnas = img.shape
for fil in range(filas):
    for col in range(columnas):
        img_gamma[fil, col] = (img[fil,col] ** gamma)

plt.imshow(img_gamma, cmap='gray')



```
