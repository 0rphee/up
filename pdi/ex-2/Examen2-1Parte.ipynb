De la carpeta `images` cargue la imagen `lena.jpg` como lo requiera para:
1.      Obtener su transformación negativa
2.      Obtener su transformación logarítmica
3.      Utilizando la transformación gamma, aclarar la imagen
4.      Utilizando la transformación gamma, oscurecer la imagen

_En todos los casos muestre la imagen que resulte de aplicar la transformación_

---

```python
import cv2 as cv
import matplotlib.pyplot as plt

original = cv.imread('../images/lena.jpg',0)
plt.imshow(original, cmap='gray')
```

```python
# transformación negativa
negativa = 255  - original
plt.imshow(negativa, cmap='gray')
```

```python
# transformación logarítmica
C = 255 / np.log(1 + np.max(original))
print(C)

s = C * np.log(1 + original)

plt.imshow(s, cmap='gray')

```

```python
# transformación gamma aclarar imagen
gamma = .001
gammaAclarado = np.power(original.copy().astype(np.float64), gamma)
plt.imshow(gammaAclarado, cmap='gray')
```

```python
# transformación gamma oscurecer imagen
gamma = 1/.1
gammaAclarado = np.power(original.copy(), gamma)
plt.imshow(gammaAclarado, cmap='gray')
```
