```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# invertir orden de canales para bgr a rgb
# [:,:,::-1]
img = cv.imread('../OpenCV-images/messi5.jpg' )[:,:,::-1]
```

```python
img.shape
```

```python
plt.imshow(img)
# si intentamos poner cmap='gray' no funciona 
# porque no es una imagen en escala de grises, 
# tiene color
```

```python
# obtener canales
# RGB
canal_r = img[:,:,0]
canal_g = img[:,:,1]
canal_b = img[:,:,2]
# cmap='Reds' para que se vea en rojo
# plt.title('Canal R') para cambiar el título de la imagen
plt.imshow(canal_r, cmap='gray') 
plt.title('Canal R')
```

```python
canal_b.shape
# ya no se muestra una tercera dimensión
# es solo una matriz de 2 dimensiones
# pero ya no hay manera de saber que color es
```

```python
plt.imshow(canal_b, cmap='gray')
```

```python
# como hacer plots con más imagenes (sublplots)
# 131, 1 fila, 3 columnas, valor1
plt.subplot(131)
plt.imshow(canal_r, cmap='gray')
plt.title('Canal R')

plt.subplot(132)
plt.imshow(canal_g, cmap='gray')
plt.title('Canal G')

plt.subplot(133)
plt.imshow(canal_b, cmap='gray')
plt.title('Canal B') 
```

```python
plt.subplot(331); plt.imshow(canal_r, cmap='gray'); plt.title('Canal R')
plt.subplot(332); plt.imshow(canal_g, cmap='gray'); plt.title('Canal G')
plt.subplot(333); plt.imshow(canal_b, cmap='gray'); plt.title('Canal B')
```

```python
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
# para mostrar canales, hay diferentes color maps
# sequenciales: cambio de saturación de un solo color incrementalmente 
# divergentes: son aquellos que tienen diferentes colores a los extremos y blanco en el centro (RdBu, PuOr, BrBG, PiYG, PRGn, coolwarm, seismic)
# ciclicos: cambian la intensidad de dos colores que se unen en el medio(hsv, twilight, twilight_shifted, hsv)
# qualitativos: miscelaneos, representan información que no importa la ordenación 
```

```python
imgRed = img.copy()
imgRed[:,:,1:] = 0
plt.imshow(imgRed)
# una imagen que no tiene color verde ni azul
# no se ve en viridis porque se mantienen las 3 dimensiones
# pero se quita la información de los otros canales
```
