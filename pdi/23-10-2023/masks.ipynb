```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../images/chicky_512.png")[:,:,::-1]
plt.imshow(img)


```

```python
mask = np.zeros_like(img)
mask[50:450, 150:450] = 255
plt.imshow(np.hstack((img, mask)))
```

```python
plt.imshow(mask & img)
```

```python
for n, i in enumerate(["r", "g", "b"]):
    histograma = cv.calcHist([img], [n], mask[:,:,0], [256], [0, 256])
    plt.plot(histograma, i)
```
