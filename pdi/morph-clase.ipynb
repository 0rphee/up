```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread("../../images/LinuxLogo.jpg", 0)
plt.imshow(img, "gray")
```

```python
UMBRAL = 128
img_bin = cv.threshold(img, UMBRAL, 255, cv.THRESH_BINARY_INV)[1]
```

```python
plt.imshow(img_bin, "gray")
```

## Erode

```python
s_e = np.ones((3,3), dtype=np.uint8)
s_e2 = np.ones((5,5), dtype=np.uint8)
```

```python
plt.imshow(np.hstack((img_bin, cv.erode(img_bin, s_e, iterations=5))), "gray")
plt.show()
```

```python
plt.imshow(np.hstack((img_bin, cv.erode(img_bin, s_e2, iterations=2))), "gray")
plt.show()
```

## Dilate

```python
plt.imshow(cv.dilate(img_bin, s_e, iterations=2), "gray")
```

```python
plt.imshow(cv.dilate(img_bin, s_e2, iterations=1), "gray")
```

```python
# It is the difference between dilation and erosion of an image.
dilation = cv.dilate(monedas_bin, s_e)
erosion = cv.erode(monedas_bin, s_e)
```

```python
plt.imshow(dilation - erosion, "gray")
```
