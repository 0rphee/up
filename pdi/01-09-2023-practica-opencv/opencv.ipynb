```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
# a partir de aquí 04-09-2023
# para transformar de BGR a RGB
# (::-1) para voltear la lista, se invierte el orden de los canales
plt.imshow(img[:,:, ::-1],)

# transformación de espacio de color
```

```python
# hay función en opencv para transformar de espacios de colores
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)  # BGR a RGB
plt.imshow(img)
```

```python
# cmap = "viridis" por defecto (para escala de grises útlil específicamente para daltónicos)
# cmap = "gray" para escala de grises, la matriz va a tener colores entre 0 y 255. 

plt.imshow(img[:,:, 0],cmap="gray")
```

```python
# acceder a pixel específico
img[0,0,0] 
```

```python
# acceder a las primeras 200 filas y 400 columnas en todos los canales
plt.imshow(img[220:400,220:400,0],cmap="gray")
```
