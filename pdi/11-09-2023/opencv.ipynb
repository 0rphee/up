```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
# imagen amarilla

canal_r  = np.full((300,300),255, dtype=np.uint8)
canal_g  = np.full((300,300),255, dtype=np.uint8)
canal_b  = np.full((300,300),0, dtype=np.uint8)

# opuesto a cv.split() est√° cv.merge()
img = cv.merge((canal_r,canal_g,canal_b))
```

```python
plt.imshow(img)
```

```python
# guardar imagen
cv.imwrite('amarillo.png',img[:,:,::-1]) # se debe invertir otra vez a bgr
```

```python
# imagen roja
canal_r2  = np.full((300,300),255, dtype=np.uint8)
canal_g2  = np.full((300,300),0, dtype=np.uint8)
canal_b2  = np.full((300,300),0, dtype=np.uint8)
roja = cv.merge((canal_r2,canal_g2,canal_b2))

plt.imshow(roja)
```

```python
cv.imwrite('roja.png',roja[:,:,::-1])
```
